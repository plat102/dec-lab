## Yêu cầu 1: Chuẩn hóa và làm sạch dữ liệu

1. Chuẩn hóa cột salary về dạng số, xử lý các giá trị như 'Thoả thuận', 'Trên X triệu', 'X - Y triệu', 'Tới X triệu', etc.
2. Tạo thêm các cột phụ: min_salary, max_salary, salary_unit (VND/USD)
3. Xử lý cột address để tách thành city và district
4. Chuẩn hóa job_title để gom nhóm các vị trí tương tự (ví dụ: 'Software Engineer', 'Developer', 'Programmer' có thể gom vào một nhóm)

```python
from helpers import extract_salary_info
```

```python
df['salary_info'] = df['salary'].apply(extract_salary_info)
df['salary_num'] = df['salary_info'].apply(lambda x: x.get('num')*(10**6)
    if (x.get('num') and 'VND' in x.get('unit')) else None)
df['min_salary'] = df['salary_info'].apply(lambda x: x.get('min')*(10**6) if x.get('min') and x.get('unit') == 'VND' else x.get('min'))
df['max_salary'] = df['salary_info'].apply(lambda x: x.get('max')*(10**6) if x.get('max') and x.get('unit') == 'VND' else x.get('max'))
df['salary_unit'] = df['salary_info'].apply(lambda  x: x.get('unit'))

df.head(10)
```

```python
from helpers import parse_address
```

```python
df['address_info'] = df['address'].apply(lambda x: parse_address(x))

# Get first address city & district info into
df['city'] = df['address_info'].apply(lambda x: x[0].get('city'))
df['district'] = df['address_info'].apply(lambda x: x[0].get('district'))
```
